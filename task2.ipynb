```python
df=pd.read_csv('train.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df_1 = df.drop_duplicates()
df_1.info()
```

```python
pclass_names = {1: 'First Class', 2: 'Second Class', 3: 'Third Class'}
df['Pclass name'] = df['Pclass'].map(pclass_names)
df.head()
```

```python
df['members'] = df['SibSp'] + df['Parch']
df.head()
```

```python
del df['Family']
```

```python
cont_val = df[['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'members']]
sns.set(style='whitegrid', palette='pastel')
sns.pairplot(cont_val, kind='reg', diag_kind='kde')
plt.show()

```

```python
sns.set(style='darkgrid')
sns.countplot(data=df, x="Pclass name", hue='Sex', palette='muted')
plt.title("Distribution of passenger's classes")
plt.show()
```

```python
sns.set_style('white')
sns.histplot(data=df, x="Age", bins=15, hue='Survived', palette=['skyblue', 'salmon'], hue_order=[1, 0], kde=True)
plt.title("Distribution by age")
plt.show()
```

```python
sns.set_style('whitegrid')
sns.histplot(data=df, x='members', bins=10, color='skyblue', kde=True)
plt.title("Distribution of passenger's family members")
plt.show()
```
